# 8.3.1 Método de Newton

**Método de Newton:** $\;$ algoritmo clásico de segundo orden que asume Hessiana definida positiva
$$\boldsymbol{\theta}_{t+1}=\boldsymbol{\theta}_t-\eta_t\mathbf{H}_t^{-1}\boldsymbol{g}_t%
\quad\text{con}\quad%
\mathbf{H}_t=\nabla^2\mathcal{L}(\boldsymbol{\theta})\rvert_{\boldsymbol{\theta}_t}%
=\nabla^2\mathcal{L}(\boldsymbol{\theta}_t)=\mathbf{H}(\boldsymbol{\theta}_t)\succ 0$$

Intuitivamente, $\mathbf{H}_t^{-1}$ "deshace" la curvatura del espacio en $\boldsymbol{\theta}_t$, facilitando así un descenso por gradiente rápido

**Algoritmo de Newton:**
> Inicializar $\boldsymbol{\theta}_0$<br>
> Para $\;t=1,2,\ldots$ (hasta convergencia):<br>
> > Evaluar $\boldsymbol{g}_t=\nabla\mathcal{L}(\boldsymbol{\theta}_t)$<br>
> > Evaluar $\mathbf{H}_t=\nabla^2\mathcal{L}(\boldsymbol{\theta}_t)$<br>
> > Hallar $\boldsymbol{d}_t$ tal que $\mathbf{H}_t\boldsymbol{d}_t=-\boldsymbol{g}_t$<br>
> > Hallar $\eta_t$ por búsqueda lineal en $\boldsymbol{d}_t$<br>
> > $\boldsymbol{\theta}_{t+1}=\boldsymbol{\theta}_t+\eta_t\boldsymbol{d}_t$

**Derivación del algoritmo de Newton:**

* Aproximación Taylor de segundo orden de $\mathcal{L}(\boldsymbol{\theta})$ en $\boldsymbol{\theta}_t$:
$$\mathcal{L}_{\text{quad}}(\boldsymbol{\theta})=\mathcal{L}(\boldsymbol{\theta}_t)+\boldsymbol{g}_t^t(\boldsymbol{\theta}-\boldsymbol{\theta}_t)+\frac{1}{2}(\boldsymbol{\theta}-\boldsymbol{\theta}_t)\mathbf{H}_t(\boldsymbol{\theta}-\boldsymbol{\theta}_t)$$
* El mínimo de $\mathcal{L}_{\text{quad}}$ se halla en:
$$\boldsymbol{\theta}=\boldsymbol{\theta}_t-\mathbf{H}_t^{-1}\boldsymbol{g}_t$$
* Así pues, si la aproximación cuadrática es buena, debemos escoger la dirección de descenso:
$$\boldsymbol{d}_t=-\mathbf{H}_t^{-1}\boldsymbol{g}_t$$
* Aunque Newton estándar usa $\eta_t=1$, conviene emplear búsqueda lineal para hallar el mejor tamaño de paso

**Ejemplo:** $\;$ Newton en 1d; $\;$ la curva sólida es $\mathcal{L}(\theta)$ y la discontinua su aproximación en $\theta_k$,
$\mathcal{L}_{\text{quad}}(\theta)$
